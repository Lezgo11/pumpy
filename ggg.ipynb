```python
from lv_before import atrial_pressure_pa, ventricular_pressure_pa
import numpy as np
```

```python
t = 0.3
T = 1  # Define the period or any other required parameter

at_p = atrial_pressure_pa(t, T)
ven_p = ventricular_pressure_pa(t, T)

import matplotlib.pyplot as plt

x = np.arange(0, 1, 0.01)
plt.plot(x, np.full_like(x, at_p), label='Atrial Pressure')
plt.plot(x, np.full_like(x, ven_p), label='Ventricular Pressure')
plt.xlabel('Time (s)')
plt.ylabel('Pressure (Pa)')
plt.title('Atrial and Ventricular Pressures Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
import csv
with open('left_heart_simple_log.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)

data = np.array(data)  # Convert to NumPy array
data[1:] = data[1:].astype(float) * 133.322  # Convert to float and multiply
print(data[0]) 
data = np.array(data[1:], dtype=float).T  # Transpose and convert to float
times = data[0, 1:10]
pressure_la = data[1, 1:10]
pressure_lv = data[2, 1:10]
vol_la = data[3, 1:10]
vol_lv = data[4, 1:10]
```

```python
#change to mmhg: 
  # Convert from Pa to mmHg
plt.figure(figsize=(10, 5))
plt.plot(times,pressure_la, label='Ventricular Pressure')
plt.plot(times,pressure_lv, label='Aortic Pressure')
plt.xlabel('Time (s)')
plt.ylabel('Pressure mmHg')
plt.title('Ventricular and Aortic Pressures Over Time')
plt.legend()
plt.grid()
plt.show()
#plt.plot(data[0], data[1], label='Atrial Pressure')

#plot volume:
plt.figure(figsize=(10, 5))
plt.plot(times,vol_la, label='Left Atrial Volume')
plt.plot(times,vol_lv, label='Left Ventricular Volume')
plt.xlabel('Time (s)')
plt.ylabel('Volume (m^3)')
plt.title('Left Atrial and Ventricular Volumes Over Time')
plt.legend()
plt.grid()
plt.show()
```
